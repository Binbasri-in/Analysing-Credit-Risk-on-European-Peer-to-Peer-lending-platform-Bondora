#  Creating Target Variables 

### **Preferred ROI**

* We weren't able to determine the procedure of handling Risk related to loan in order to determine **Preferred ROI**.

* In order to complete the task in hand and complete it, we'll calculate **ROI** instead.
>**ROI** = Investment Gain / Investment Base
>
> **ROI** = Amount lended * interest/100

* **InterestAndPenaltyDebtServicingCost**	Service cost related to the recovery of the debt based on the interest and penalties of the investment

* **InterestAndPenaltyWriteOffs**	Interest that was written off on the investment

* **PrincipalDebtServicingCost**	Service cost related to the recovery of the debt based on the principal of the investment

* **PrincipalWriteOffs**	Principal that was written off on the investment

* **PurchasePrice**	Investment amount or secondary market purchase price

```python
loan_data_temp = loan_data[['Amount', 'Interest']]
loan_data_temp['InterestAmount'] = (loan_data_temp['Amount']*(loan_data_temp['Interest']/100))
loan_data_temp['TotalAmount'] = (loan_data_temp['InterestAmount'] + loan_data_temp['Amount'])
loan_data_temp['ROI'] = (loan_data_temp['InterestAmount'] / loan_data_temp['TotalAmount'])*100
loan_data['ROI'] = loan_data_temp['ROI']
```

```python
loan_data_temp.head()
```

## 1.1. EMI:
- Flat-Rate: 
(PrincipleAmount + (PrincipleAmount  x  # of Years  x  InterestRate)) / (# of Years / Months)
- **Reducing-Balance:**
__<center>P * [( r * (1 + r)^n)) / ((1 + r)^n - 1)]__  </center>   
where:   
P = Princiapl amount borrowed -> "Amount"  
r = Periodic monthly interest rate  ->  "Interest"  
n = Total number of monthly payments  ->  "LoanDuration"  

```python
Emi_Required_features = loan_data[['Amount','Interest','LoanDuration']]
Emi_Required_features.head()
```

```python
for ind, row in Emi_Required_features.iterrows():
    Emi_Required_features.loc[ind,'EMI'] = row['Amount'] * row['Interest'] * (np.power((1 + row['Interest']),row['LoanDuration'])) / (np.power((1 + row['Interest']),row['LoanDuration']) - 1)
```

```python
Emi_Required_features.head()
```

### Eligible Loan Amount

```python
ELA_required_features = loan_data[['IncomeTotal','LiabilitiesTotal','Interest','LoanDuration','DebtToIncome','CreditScoreEsMicroL','MonthlyPayment']]
```

```python
ELA_required_features['Surplus Amount'] = ELA_required_features['IncomeTotal']-ELA_required_features['LiabilitiesTotal']
```

```python
ELA_required_features['Avg_Inc'] = ELA_required_features['Surplus Amount'] * 0.3
ELA_required_features['Total_Loan_Amnt'] = np.round((loan_data['AppliedAmount'] + (loan_data['AppliedAmount'] * loan_data['Interest']) /100)*loan_data['LoanDuration'])
ELA_required_features.head()
```

```python
def eligible_loan_amnt(ELA_required_features):
    Ava_Inc = ELA_required_features['Avg_Inc'].values
    Total_Loan_Amnt = ELA_required_features['Total_Loan_Amnt'].values
    ELA = np.empty(0)
    for i in range(len(Ava_Inc)):
        if Total_Loan_Amnt[i] <= Ava_Inc[i]:
            ELA = np.append(ELA, Total_Loan_Amnt[i])
        else:
            ELA = np.append(ELA, Ava_Inc[i])
    return ELA
```

```python
ELA_required_features['ELA'] = eligible_loan_amnt(ELA_required_features)
```

```python
ELA_required_features.head()
```

### handling outliers

```python
# Finding the outliers 
def find_outliers_IQR(df, cols):
    for i in cols:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        IQR=q3-q1
        outliers = df[((df[i]<(q1-1.5*IQR)) | (df[i]>(q3+1.5*IQR)))]
    return outliers
```
